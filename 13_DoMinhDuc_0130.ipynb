```python
import pandas as pd
import numpy as np
import sqlite3 
```

```python
#1.
class NhanVien():
    def __init__(self, ma_nv, luong_cb):
        self.__ma_nv = str(ma_nv)
        self.__luong_cb = float(luong_cb)

    def get_ma_nv(self):
        return self.__ma_nv
    
    def get_luong_cb(self):
        return self.__luong_cb
    
    def tinh_luong(self):
        return self.__luong_cb

#2.
class NhanVienTangCa(NhanVien):
    def tinh_luong(self):
        return super().tinh_luong() * 1.2
    
class NhanVienChuyenNghiep(NhanVien):
    def tinh_luong(self):
        return super().tinh_luong()+ 500000
    
#3.
if __name__ == "__main__":
    luong_base = 10000000
    ds_nhan_vien  = [ 
        NhanVien("NV001", luong_base),
        NhanVienTangCa("NV002", luong_base),
        NhanVienChuyenNghiep("NV003", luong_base)
        ]
    
    print(f"{'Mã NV':<10} | {'Tiền lương'}")
    for nv in ds_nhan_vien:
        print(f"{nv.get_ma_nv():<10} | {nv.tinh_luong():,.0f} VND")
```

```python
san_luong = [12, None, 8, 15, None, 10]

arr = np.array(san_luong, dtype=float).reshape(2, 3)

mean_value = np.nanmean(arr)
arr = np.nan_to_num(arr, nan=mean_value)

print("Mảng sau khi xử lý:")
print(arr)

lon_hon_9 = arr[arr > 9]
print("Các phần tử > 9:", lon_hon_9)
print("Giá trị nhỏ nhất:", lon_hon_9.min())
```

```python
df = pd.read_csv("../DATA/nhan_su.csv")

df_loc = df[
    df["Ma_NV"].str.contains("NS", case=False, na=False) &
    (df["Nam_Cong_Tac"] >= 2021)
].copy()

mean_tt = df_loc["Thanh_Tich"].mean()
df_loc["Thanh_Tich"] = df_loc["Thanh_Tich"].fillna(mean_tt)

df_loc
```

Câu 2 c


```python
tong_thanh_tich = df_loc.groupby("Phong_Ban")["Thanh_Tich"].sum()
tong_thanh_tich
```

Câu 2 d

```python
pivot = pd.pivot_table(
    df_loc,
    index="Phong_Ban",
    columns="Nam_Cong_Tac",
    values="Thanh_Tich",
    aggfunc="sum",
    fill_value=0
)

pivot
```

Câu 2 e

```python
conn = sqlite3.connect("../OUTPUT/KetQuaPhanTich.db")

pivot.to_sql("bang_luong_2024", conn, if_exists="replace", index=True)

kiem_tra = pd.read_sql(
    "SELECT * FROM bang_luong_2024 LIMIT 5",
    conn
)

conn.close()

kiem_tra
```
